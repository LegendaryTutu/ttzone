```python
#导入包
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#中文显示
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"]=False

#使用pandas读取数据支持xls和xlsx
data=pd.read_excel("C:\\Users\\有福有德\\Desktop\\上海统计分析python20200415\\data\\烟草.xlsx")
data.head(6)
```

```python
#相关分析corr
plt.scatter(x="总耗",y="产量",s=100,data=data,alpha=0.3)#alpha=气泡透明度，22为尺寸
raw=data.iloc[:,[1,2,3]]
raw.corr(method="spearman")#显示所有变量，method=pearson、kendall、spearman
# raw.corr()["总耗"]#显示绩效总分与其他变量的相关系数
#raw["绩效总分"].corr(raw["适应总分"])#只显示2个变量的相关分析

# from scipy import stats
# r,p=stats.pearsonr(raw["ed"],raw["income"])#显示p值
# print(r,p)
```

```python

from scipy import stats
#卡方配合度检验
#pandas数据
# obs=data.iloc[:,3]
# freq=obs.value_counts(normalize=True)#normalize=True与期望exp小数对应
# exp=[.4,.3,.3]
# stats.chisquare(f_obs=freq,f_exp=exp,axis=0)

from scipy.stats import chi2_contingency
#独立性卡方检验
d=pd.crosstab(data["性别"],data["部门"],margins=True,normalize=1)#注意normalize选项会影响卡方的计算
              #normalize="all" |"index"or0|"columus"or1总体%、行%、列%
print(d)
x2,p,df,exp=chi2_contingency(d)
print("x2,p:%.2f,%.2f" %(x2,p))
```
