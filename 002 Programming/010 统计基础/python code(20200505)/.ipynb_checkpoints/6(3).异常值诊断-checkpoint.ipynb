```python
#导入包
import numpy as np
import pandas as pd
import matplotlib as plt
import statsmodels.formula.api as smf
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"]=False

data=pd.read_excel("C:\\Users\\有福有德\\Desktop\\上海统计分析python20191216\\data\\短期绩效.xlsx")
#回归方法
result=smf.ols("绩效总分~适应总分+情绪总分",data=data).fit()#np.log(绩效总分)
print(result.summary(alpha=0.05))
```

```python
from statsmodels.stats.outliers_influence import OLSInfluence
outlier=result.get_influence()#hat_matrix_diag杠杆(2倍均值);resid_studentized_external学生化残差(2或3);cooks_distance界值4/（n-k-1）
out=outlier.summary_frame()#汇总所有异常指标
out.head(6)

# 绘制变量间的散点图
s=out.iloc[:,[3,5,7]]
sns.pairplot(s,diag_kind='kde',kind='reg')#diag_kind='hist'|'kde'
```

```python
out=result.get_influence().summary_frame()#汇总所有异常指标
out1=out.iloc[:,[3,5,7]]#单独提取cook,h,残差指标
data1=data
data1[["cook","hat","stud_resid"]]=out1#复制保存
data2=data1[np.abs(data1["stud_resid"])<2][data1["cook"]<(4/(100-2-1))]#[data1["hat"]<2*np.mean(data1.hat)]#设置异常条件
print(data2.shape)

result1=smf.ols("绩效总分~适应总分+情绪总分",data=data2).fit()
print(result1.summary(alpha=0.05))
out2=result1.get_influence().summary_frame()
sns.jointplot(x=result1.fittedvalues,y=out2.student_resid,data=data2,kind="reg")#"reg"|"resid"|"kde"|"hex"
```
