```python
#导入包
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn import linear_model
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#使用pandas读取数据支持xls和xlsx
data=pd.read_excel("C:\\Users\\有福有德\\Desktop\\上海统计分析python20200418\\data\\bankloan_binning.xlsx")
data.head(6)#显示数据开头6行
```

```python
#===========电商dm特征筛选功能=============

#第1步：业务驱动：业务指导
#第2步：数据驱动：y与x的回归+相关
#第3步：业务+数据驱动：x与x相关+人工字段
#第4步：数据+业务驱动：logistic+递归式特征消除(RFE)
#第5步：主成分分析：线性或非线性
#----------------------

#第1步：业务指导
#删除冗余变量，并确定重要的变量ed
data1=data.drop(["ID","age_group"],axis=1)
data1.head(6)
```

```python
#第2步：数据驱动：y与x的回归+相关
corr=data1.corr()["default"].plot.bar()#检查最小相关plot.line、box，并以0.1相关与删除30%变量为标准
print("第2步：",corr)

data2=data1.drop(["income"],axis=1)
data2.head(6)
```

```python
#第3步：x与x相关分析
data2.iloc[:,0:7].corr(method="spearman")

# data2.plot.scatter(x="debtinc",y="othdebt")
dt=data2[data2["othdebt"]<=300]
dt.plot.scatter(x="othdebt",y="debtinc")

# #拟合线性形式的模型
from scipy.optimize import curve_fit

def f(x,b0,b1):
    return b0+b1*x     #调整x和y的任意函数关系,如b0*np.exp(b1*dt["othdebt"])
popt,pcov=curve_fit(f,dt["othdebt"],dt["debtinc"])
b0=popt[0]
b1=popt[1]

data2["comb"]=b0+b1*dt["othdebt"]#整合新字段并计算r方
# print("r**2:",(data2["comb"].corr(data2["y"]))**2)
data3=data2.drop(["othdebt","debtinc"],axis=1)
data3.head(6)

# 线性：方程为 Y = b0 + (b1 * x) 的模型。
# 二次：方程为 Y = b0 + (b1 * x) + (b2 * x**2) 的模型。
# 对数：方程为 Y = b0 + (b1 * ln(x)) 的模型。
# 指数：方程为 Y = b0 * (e**(b1 * x)) or ln(Y) = ln(b0) + (b1 * x) 的模型。
# 逆：方程为 Y = b0 + (b1 / x) 的模型。
# 幂：方程式为 Y = b0 * (x**b1) 或 ln(Y) = ln(b0) + (b1 * ln(x)) 的模型。
# S 曲线：方程式为 Y = e**(b0 + (b1/x)) 或 ln(Y) = b0 + (b1/x) 的模型。
# Logistic：方程为 Y = 1 / (1/u + (b0 * (b1**x))) 的模型,其中u为上限值(需要大于y的值)。
```

```python
# 第4步：logistic回归——RFE
from sklearn.feature_selection import RFE

X=data3.iloc[:,[0,2,3,4]]
y=data3.iloc[:,5]
fit=linear_model.LogisticRegression()
selector = RFE(fit,n_features_to_select=2).fit(X, y)#递归式特征消除:Recursive feature elimination(RFE)
print("第4步：",selector.support_)

#其他特征选择技术feature_selection中SelectFromModel,并可以定义任意模型
# from sklearn.pipeline import Pipeline
# from sklearn.feature_selection import SelectFromModel
# from sklearn.svm import LinearSVC
# from sklearn.ensemble import RandomForestClassifier

# clf = Pipeline([
#                ("特征选择",SelectFromModel(LinearSVC(penalty="l2",random_state=0))),
#                ("模型",RandomForestClassifier(oob_score=True,random_state=0))
#                ])
# rfit=clf.fit(X,y)
# print(rfit.named_steps['特征选择'].get_support())#特征选择
# print(rfit.named_steps['模型'].feature_importances_)#特征重要性
# print(rfit.named_steps['模型'].oob_score_)#随机森林的r2值
#第5步：主成分分析
```
