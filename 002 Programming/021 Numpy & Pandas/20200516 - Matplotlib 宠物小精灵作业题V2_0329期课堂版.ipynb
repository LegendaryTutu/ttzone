## 折线图

绘制这两个函数的图像,到一张图片上面

横坐标数据范围从0.01 - 10




图片显示范围,横坐标从(-1 , 15)

纵坐标从(-10,20)

分别为两张图片添加图例说说明来自哪一个函数

图片标题为: <两个常见函数>

字体大小都为20

```python
def f(x):
    return np.log(x) - x
```

```python
def f2(x):
    return x*np.log(x)
```

```python
x = np.linspace(0.01,10,100)
y = f(x)
```

```python
y2 = f2(x)
```

```python
plt.plot(x,y,label ='y=lnx - x' )
plt.plot(x,y2,label = "y=xlnx")
plt.xlim(-1,15)
plt.ylim(-10,20)
plt.legend(fontsize=20)
plt.title('两个常见函数',fontsize=20);
```
