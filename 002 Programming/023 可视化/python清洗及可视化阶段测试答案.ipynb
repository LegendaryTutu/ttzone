### 问题1：对大陆这一列中的空值进行填充

简写：[AF]非洲, [EU]欧洲, [AS]亚洲, [OA]大洋洲, [NA]北美洲, [SA]南美洲, [AN]南极洲。

```python
# 先查看一下各洲简写字母
drinks.continent.value_counts()
```

```python
continent_dict = {"AS":"亚洲",'EU':'欧洲','AF':"非洲",'SA':'南美洲','OC':'大洋洲'}
```

```python
drinks['大洲'] = drinks.continent.replace(continent_dict)
```

```python
# 空值应该是北美洲,进行填充
drinks.大洲.fillna('北美洲',inplace=True)
```

```python
drinks
```

### 问题2：打印出每个大陆(continent)的红酒消耗(wine_servings)的描述性统计值

```python
# 运行以下代码
drinks.groupby('大洲').wine_servings.describe()
```

### 问题3：哪个大陆(continent)平均消耗的啤酒(beer)更多？

```python
# 运行以下代码
drinks.groupby('大洲').beer_servings.mean()
```

###  问题4：打印出每个大陆每种酒类别的消耗平均值

```python
# 运行以下代码
drinks.groupby('大洲').mean()
```
