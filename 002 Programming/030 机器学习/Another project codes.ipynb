```python
# draw three stocks in one chart

def stocks_data(symbols):
    L = []
    for symbol in symbols:
        mic = pd.read_csv("{}.csv".format(symbol),index_col='Date',
                parse_dates=True, usecols=['Date', 'Close'], na_values=['nan'])
        mic = mic.rename(columns={'Close': symbol})
        L.append(mic)
    return L
symbols = ['GOOG','IBM','MSFT ']
L = stocks_data(symbols)
df = pd.concat(L, axis=1, join='inner')

plt.figure(figsize=(8,6), dpi=70)
for i in symbols:
    plt.plot(df.index, df[i], label=i);
plt.legend();
```

```python
# Calculation of Relative Strength Index
def rsi(values):
    up = values[values>0].mean()
    down = -1*values[values<0].mean()
    return 100 * up / (up + down)
```

```python
# Momentum 
for stock in range(len(data)):     # stock refers to row while close columns
    data[stock]['Momentum'] = (data[stock]['Close']-data[stock]['Close'].shift(1)).fillna(0)
    data[stock]['RSIC'] = data[stock]['Momentum'].rolling(center=False, window=2).apply(rsi).fillna(0)
data[:].tail(5)    #  question !
```

```python
def KAMA(price, n=10, pow1=2, pow2=30):

    Diffx = abs(price - price.shift(1) )  

    ER_num = abs( price - price.shift(n) )
    ER_den = absDiffx.rolling(window=n,center=False).sum()
    ER = ER_num / ER_den

    sc = ( ER*(2.0/(pow1+1)-2.0/(pow2+1.0))+2/(pow2+1.0) ) ** 2.0


    answer = np.zeros(sc.size)
    N = len(answer)
    first_value = True

    for i in range(N):
        if sc[i] != sc[i]:
            answer[i] = np.nan
        else:
            if first_value:
                answer[i] = price[i]
                first_value = False
            else:
                answer[i] = answer[i-1] + sc[i] * (price[i] - answer[i-1])
    return answer
```
